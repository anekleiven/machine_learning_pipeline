## Import libraries

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

## Convert boolean columns to numeric (0/1)

```python
# Convert boolean columns to numeric
# 1 = True, 0 = False
boolean_cols = ['has_gnomAD_AF', 'In_Hotspot', 'IN_DOMAIN', 'IN_FUNC_SITE']
for col in boolean_cols: 
  df_ml[col] = df_ml[col].astype(int) 
```
